```r
train$Sex <- revalue(train$Sex, c("female"=0))
train$Sex <- revalue(train$Sex, c("male"=1))
```

```r
test$Sex <- revalue(test$Sex, c("female"=0))
test$Sex <- revalue(test$Sex, c("male"=1))
```

```r
is.na(train$Age) <- mean(train$Age)
is.na(train$Fare) <- mean(train$Fare)
is.na(test$Age) <- mean(test$Age)
is.na(test$Fare) <- mean(test$Fare)
```

```r
model <- glm(Survived~ Pclass+Sex+Age,data=train,family=binomial)
```

```r
summary(model)
```

```r
round(predict(model,test[c('Pclass','Sex','Age')],type='response'),0)
```

```r
test_pred <- matrix(c(1,1,20),ncol=3)
colnames(test_pred) <- c("Pclass","Sex","Age")
test_pred <- data.frame(test_pred)
test_pred$Sex <- as.factor(test_pred$Sex)
```

```r
round(predict(model,test_pred,type='response'),0)
```
